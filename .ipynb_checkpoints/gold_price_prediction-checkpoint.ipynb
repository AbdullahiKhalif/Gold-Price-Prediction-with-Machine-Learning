
# Gold Price Prediction

The **Gold Price Prediction** typically refers to a project aimed at developing a model or system that can forecast the future prices of gold.

-------------------------

**The dataset contains the following variables:**

* **'1.	Date':** This variable represents the specific trading day, acting as a unique identifier for each record. It's essential for organizing and sorting the data chronologically, which is crucial for time series analysis.

* **'2.	Open':** This variable represents the price of gold at the start of the trading day, measured in USD per gram. It indicates the initial market valuation and sentiment for the day, serving as an important metric for analyzing daily price trends and forecasting future opening prices.

* **'3.	High':** This variable denotes the highest price of gold recorded during the trading day, measured in USD per gram. It provides insight into the peak market valuation and potential resistance levels within the day's trading range.

* **'4.	Low':** This variable indicates the lowest price of gold recorded during the trading day, measured in USD per gram. It reflects the minimum market valuation and potential support levels, helping to understand the volatility and trading range for the day.

* **'5.	Close':** This variable represents the price of gold at the end of the trading day, measured in USD per gram. It is a critical metric for analysis and forecasting, as it reflects the final market sentiment and overall stability at the close of the market.

* **'6.	Volume':** This variable shows the total number of gold grams traded during the trading day. It indicates the level of market activity and liquidity, with higher volumes suggesting strong market interest and potential significant price movements.

* **'7.	Currency':** This variable specifies the currency in which the gold prices are quoted, typically USD (United States Dollar). It ensures clarity and consistency in the financial metrics presented, facilitating accurate international comparisons and conversions.

### Checking Outlier if exist

```python
# Define a function to detect outliers based on the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)].shape[0]

# Columns to check for outliers
columns_to_check = ['Open', 'High', 'Low', 'Close', 'Volume']

# Calculate the number of outliers in each column
outliers_count = {column: detect_outliers_iqr(df, column) for column in columns_to_check}
outliers_count
```

-------------------------

### Handling outliers by droping permenentally

```python
# Define a function to remove outliers based on the IQR method for a given column
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply the outlier removal for each column individually
df = df.copy()
for column in columns_to_check:
    df = remove_outliers_iqr(df, column)

# Check the new shape of the dataset to see how many records remain
df.shape
```

# Building Machine Learning Models

### Separate features (X) and target variable (y)

```python
# Splitting features and target
X = df[['Open', 'High', 'Low', 'Volume']] # All Varibles are our feature eccept Close
y = df['Close']  # Target variable, ''Close' mean Last trand
```

### Train Test Split

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## 1. Linear Regression Model

```python
lin_reg_model = LinearRegression()
```

### Train the model on the training set

```python
lin_reg_model.fit(X_train, y_train)
```

```python
print("Linear Score Average Trained: ", lin_reg_model.score(X_train, y_train))
```

```python
print("Linear Score Average Tested: ", lin_reg_model.score(X_test, y_test))
```

```python
# Create a DataFrame to display the evaluation metrics in a table format
evaluation_metrics = pd.DataFrame({
    "Metric": ["Mean Squared Error (MSE)", "Coefficient of Determination (R²)"],
    "Value": [19.0437, 0.999930]
})

evaluation_metrics
```

```python
predictions = lin_reg_model.predict(X_test)
plt.title('Linear Regression Model Accuracy')
plt.scatter(y_test,predictions)
plt.savefig('visualizations/lin_reg.png')  # Save the image
```
